#–ó–∞–¥–∞—á–∞ 1
–ù–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é –Ω–∞ –≤—Ö–æ–¥ –∫–æ—Ç–æ—Ä–æ–π –ø–æ–¥–∞–µ—Ç—Å—è —Å—Ç—Ä–æ–∫–∞, —Å–æ—Å—Ç–æ—è—â–∞—è –∏–∑ –ª–∞—Ç–∏–Ω—Å–∫–∏—Ö –±—É–∫–≤.
–§—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –≤–µ—Ä–Ω—É—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≥–ª–∞—Å–Ω—ã—Ö –±—É–∫–≤ (a, e, i, o, u) –≤ —ç—Ç–æ–π —Å—Ç—Ä–æ–∫–µ. –ù–∞–ø–∏—Å–∞—Ç—å —Ç–µ—Å—Ç—ã –¥–ª—è –∫–æ–¥–∞

```python
def count_vowels(s: str) -> int:
    return sum(1 for letter in s.lower() if letter in "aeiou")

# –¢–µ—Å—Ç—ã
def test_count_vowels():
    assert count_vowels('hello') == 2
    assert count_vowels('rhythm') == 0
    assert count_vowels('AEIOU') == 5
    assert count_vowels('abcdefghijklmnopqrstuvwxyz') == 5
    print("Succes!")

if __name__ == "__main__":
    test_count_vowels()
```

# –ó–∞–¥–∞—á–∞ 2
–ù–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é –Ω–∞ –≤—Ö–æ–¥, –∫–æ—Ç–æ—Ä–æ–π –ø–æ–¥–∞–µ—Ç—Å—è —Å—Ç—Ä–æ–∫–∞. –§—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –≤–µ—Ä–Ω—É—Ç—å true, –µ—Å–ª–∏
–∫–∞–∂–¥—ã–π —Å–∏–º–≤–æ–ª –≤ —Å—Ç—Ä–æ–∫–µ –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è —Ç–æ–ª—å–∫–æ 1 —Ä–∞–∑, –∏–Ω–∞—á–µ –¥–æ–ª–∂–Ω–∞ –≤–µ—Ä–Ω—É—Ç—å false. –ù–∞–ø–∏—Å–∞—Ç—å —Ç–µ—Å—Ç—ã –¥–ª—è –∫–æ–¥–∞

```python
def is_letter_unique(s: str) -> bool:
    return len(s) == len(set(s))

# –¢–µ—Å—Ç—ã
def test_all_unique():
    assert is_letter_unique('abc') == True
    assert is_letter_unique('aabc') == False
    assert is_letter_unique('') == True
    assert is_letter_unique('üôÇüôÉ') == True
    print("Succes!")

if __name__ == "__main__":
    test_all_unique()
```

# –ó–∞–¥–∞—á–∞ 3
–ù–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ —á–∏—Å–ª–æ –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ
–±–∏—Ç —Ä–∞–≤–Ω—ã—Ö 1 –≤ —ç—Ç–æ–º —á–∏—Å–ª–µ. –ù–∞–ø–∏—Å–∞—Ç—å —Ç–µ—Å—Ç—ã –¥–ª—è –∫–æ–¥–∞

```python
def bit_count(n: int) -> int:
    return bin(n).count('1')

# –¢–µ—Å—Ç—ã
def test_bit_count():
     assert bit_count(0) == 0
     assert bit_count(1) == 1
     assert bit_count(7) == 3
     assert bit_count(2**10 + 2**3) == 2
     print("Succes!")

if __name__ == "__main__":
    test_bit_count()  
```

# –ó–∞–¥–∞—á–∞ 4
–ù–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ —á–∏—Å–ª–æ. –§—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –≤–µ—Ä–Ω—É—Ç—å —Ç–æ,
—Å–∫–æ–ª—å–∫–æ —Ä–∞–∑ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –ø–µ—Ä–µ–º–Ω–æ–∂–∞—Ç—å —Ü–∏—Ñ—Ä—ã —á–∏—Å–ª–∞ –∏–ª–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø–µ—Ä–µ–º–Ω–æ–∂–µ–Ω–∏—è, —á—Ç–æ–±—ã
–ø–æ–ª—É—á–∏–ª–æ—Å—å —á–∏—Å–ª–æ —Å–æ—Å—Ç–æ—è—â–µ–µ –∏–∑ –æ–¥–Ω–æ–π —Ü–∏—Ñ—Ä—ã.
–ù–∞–ø—Ä–∏–º–µ—Ä, –¥–ª—è –≤—Ö–æ–¥–Ω–æ–≥–æ —á–∏—Å–ª–∞:
¬∑ 39 —Ñ—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –≤–µ—Ä–Ω—É—Ç—å 3, —Ç–∞–∫ –∫–∞–∫ 3*9=27 => 2*7=14 => 1*4=4
¬∑ 4 —Ñ—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –≤–µ—Ä–Ω—É—Ç—å 0, —Ç–∞–∫ –∫–∞–∫ —á–∏—Å–ª–æ —É–∂–µ —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ –æ–¥–Ω–æ–π —Ü–∏—Ñ—Ä—ã
¬∑ 999 —Ñ—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –≤–µ—Ä–Ω—É—Ç—å 4, —Ç–∞–∫ –∫–∞–∫ 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. –ù–∞–ø–∏—Å–∞—Ç—å —Ç–µ—Å—Ç—ã –¥–ª—è –∫–æ–¥–∞

```python
def mult_numbers(n: int) -> int:
    res = 1
    for num in str(n):
        res *= int(num)
    return res


def multiplicative_persistence(n: int) -> int:
    count = 0
    while n >= 10:
        n = mult_numbers(n)
        count += 1
    return count

# –¢–µ—Å—Ç—ã
def test_multiplicative_persistence():
     assert multiplicative_persistence(39) == 3
     assert multiplicative_persistence(4) == 0
     assert multiplicative_persistence(999) == 4
     assert multiplicative_persistence(10) == 1
     print("Succes!")

if __name__ == "__main__":
    test_multiplicative_persistence()  
```

# –ó–∞–¥–∞—á–∞ 5
–ù–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –¥–≤–∞ —Ü–µ–ª–æ—á–∏—Å–ª–µ–Ω–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–∞ –æ–¥–∏–Ω–∞–∫–æ–≤–æ–π –¥–ª–∏–Ω—ã –∏
–≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –¥–≤—É—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤. –ù–∞–ø–∏—Å–∞—Ç—å —Ç–µ—Å—Ç—ã –¥–ª—è –∫–æ–¥–∞

```python
def mse(pred, true):
    return sum((p - t) ** 2 for p, t in zip(pred, true)) / len(pred)

# –¢–µ—Å—Ç—ã
def test_mse():
    assert mse([1, 2, 3], [1, 2, 3]) == 0
    assert mse([1, 2, 3], [2, 3, 4]) == 1
    assert mse([0, 0, 0], [1, 1, 1]) == 1
    assert mse([1, 2], [3, 4]) == 4
    assert round(mse([1, 2, 3], [1, 3, 5]), 2) == 1.67
    print("Succes!")

if __name__ == "__main__":
    test_mse()
```

# –ó–∞–¥–∞—á–∞ 6
–ù–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é, –ø—Ä–∏–Ω–∏–º–∞—é—â–∞—è —Ü–µ–ª–æ–µ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ —á–∏—Å–ª–æ. –§—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –≤–µ—Ä–Ω—É—Ç—å
—Å—Ç—Ä–æ–∫—É –≤–∏–¥–∞ ‚Äú(n1**p1)(n2**p2)‚Ä¶(nk**pk)‚Äù –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—é—â–∞—è —Ä–∞–∑–ª–æ–∂–µ–Ω–∏–µ —á–∏—Å–ª–∞ –Ω–∞ –ø—Ä–æ—Å—Ç—ã–µ
–º–Ω–æ–∂–∏—Ç–µ–ª–∏ (–µ—Å–ª–∏ pi == 1, —Ç–æ –≤—ã–≤–æ–¥–∏—Ç—å —Ç–æ–ª—å–∫–æ ni).
–ù–∞–ø—Ä–∏–º–µ—Ä, –¥–ª—è —á–∏—Å–ª–∞ 86240 —Ñ—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –≤–µ—Ä–Ω—É—Ç—å ‚Äú(2**5)(5)(7**2)(11)‚Äù. –ù–∞–ø–∏—Å–∞—Ç—å —Ç–µ—Å—Ç—ã –¥–ª—è –∫–æ–¥–∞

```python
def prime_decomposition(n: int) -> str:
    if n == 1:
        return ''
    factors = []
    d = 2
    while d * d <= n:
        cnt = 0
        while n % d == 0:
            n //= d
            cnt += 1
        if cnt:
            factors.append((d, cnt))
        d += 1 if d == 2 else 2
    if n > 1:
        factors.append((n, 1))

    parts = []
    for p, e in factors:
        if e == 1:
            parts.append(f"({p})")
        else:
            parts.append(f"({p}**{e})")
    return ''.join(parts)

# –¢–µ—Å—Ç—ã
def test_prime_decomposition():
    assert prime_decomposition(86240) == '(2**5)(5)(7**2)(11)'
    assert prime_decomposition(1) == ''
    assert prime_decomposition(2) == '(2)'
    assert prime_decomposition(12) == '(2**2)(3)'
    print("Succes!")

if __name__ == "__main__":
    test_prime_decomposition()  
```

# –ó–∞–¥–∞—á–∞ 7
–ù–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é, –ø—Ä–∏–Ω–∏–º–∞—é—â–∞—è —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ n, –∑–∞–¥–∞—é—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫—É–±–∏–∫–æ–≤. –§—É–Ω–∫—Ü–∏—è
–¥–æ–ª–∂–Ω–∞ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å, –º–æ–∂–Ω–æ –ª–∏ –∏–∑ –¥–∞–Ω–Ω–æ–≥–æ –∫–æ–ª-–≤–∞ –∫—É–±–∏–∫–æ–≤ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –ø–∏—Ä–∞–º–∏–¥—É, —Ç–æ –µ—Å—Ç—å
–º–æ–∂–Ω–æ –ª–∏ –ø—Ä–µ–¥—Å—Ç–∞–≤–∏—Ç—å —á–∏—Å–ª–æ n –∫–∞–∫ 1^2+2^2+3^2+‚Ä¶+k^2. –ï—Å–ª–∏ –º–æ–∂–Ω–æ, —Ç–æ —Ñ—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞
–≤–µ—Ä–Ω—É—Ç—å k, –∏–Ω–∞—á–µ —Å—Ç—Ä–æ–∫—É ‚ÄúIt is impossible‚Äù. –ù–∞–ø–∏—Å–∞—Ç—å —Ç–µ—Å—Ç—ã –¥–ª—è –∫–æ–¥–∞

```python
def pyramid(n: int):
    count = 1
    res = 1
    while n != count:
        res += 1
        count += res ** 2
        if count > n:
            return "It is impossible"
    return res

# –¢–µ—Å—Ç—ã
def test_pyramid():
    assert pyramid(1) == 1
    assert pyramid(5) == 2
    assert pyramid(14) == 3
    assert pyramid(32) == "It is impossible"
    print("Succes!")


if __name__ == "__main__":
    test_pyramid()  
```

# –ó–∞–¥–∞—á–∞ 8
–§—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ —á–∏—Å–ª–æ –∏ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç —è–≤–ª—è–µ—Ç—Å—è –ª–∏ –æ–Ω–æ —Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–º, —Ç.–µ. —Å—É–º–º–∞ —Ü–∏—Ñ—Ä –¥–æ —Å—Ä–µ–¥–Ω–∏—Ö —Ä–∞–≤–Ω–∞ —Å—É–º–º–µ —Ü–∏—Ñ—Ä –ø–æ—Å–ª–µ. –°—Ä–µ–¥–Ω–∏–º–∏ –≤ —Å–ª—É—á–∞–µ –Ω–µ—á–µ—Ç–Ω–æ–≥–æ —á–∏—Å–ª–∞ —Ü–∏—Ñ—Ä —Å—á–∏—Ç–∞—Ç—å –æ–¥–Ω—É —Ü–∏—Ñ—Ä—É, –≤ —Å–ª—É—á–∞–µ —á–µ—Ç–Ω–æ–≥–æ - –¥–≤–µ —Å—Ä–µ–¥–Ω–∏—Ö. –ù–∞–ø–∏—Å–∞—Ç—å —Ç–µ—Å—Ç—ã –¥–ª—è –∫–æ–¥–∞

```python
def is_balanced_number(n: int) -> bool:
    if len(str(n)) % 2 == 0:
        left_mid = len(str(n)) // 2
        right_mid = left_mid + 1
        return sum(int(i) for i in str(n)[:left_mid]) == sum(int(i) for i in str(n)[right_mid + 1:])
    else:
        mid = (len(str(n)) - 1) // 2
        return sum(int(i) for i in str(n)[:mid]) == sum(int(i) for i in str(n)[mid + 1:])

# –¢–µ—Å—Ç—ã
def test_is_balanced_number():
    assert is_balanced_number(121) == True
    assert is_balanced_number(12321) == True
    assert is_balanced_number(1234007) == False
    print("Succes!")

if __name__ == "__main__":
    test_is_balanced_number() 
```
